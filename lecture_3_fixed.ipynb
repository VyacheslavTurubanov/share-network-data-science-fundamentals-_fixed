# Section 1: Introduction to Pandas Series

In this section, we'll be covering the **Pandas Series**, which is the Pandas equivalent of a column of data that we'd see in SQL or Excel. 

### What is a Series?
Pandas Series are really just list with additional features layered into them to make them easier to work with. Almost every piece of data in a Pandas data structure has a NumPy array hiding underneath it.

**Key Differences:**
* **Custom Index:** Unlike list, we can give a Series a custom index.
* **Optional Name:** Analogous to a column header or title.
* **Collection:** Two or more Series grouped together form a **Pandas DataFrame** (a table).

**Instructor Tip:** Generally, stick to Base Python operators unless you need the specific advantages of a pandas method (like handling missing data) or are chaining multiple operations together.

### Handling Missing Data in Arithmetic
One of the biggest advantages of using methods like `.add()` over the `+` operator is the **`fill_value`** argument. By default, any operation involving a missing value (`NaN`) will return `NaN`. Using `fill_value` allows you to impute a value (like 0) before the calculation.

```python
# Create a series with a missing value (NaN)
my_series = pd.Series([1, np.nan, 2, 3, 4], name="Numeric Series")

# Standard addition (Missing values stay missing)
print("Standard Addition:")
print(my_series + 1)

# Addition with fill_value (Treats NaN as 0 before adding)
print("\nAddition with fill_value=0:")
print(my_series.add(1, fill_value=0))
```

### Series vs. Series Operations
You can perform arithmetic between two Series. They must be of compatible lengths for broadcasting. If one Series has a missing value at an index and the other does not, the result will be `NaN` unless `fill_value` is used.

```python
# Second series for comparison
my_series2 = my_series.fillna(0).astype("int64")

# Adding two series
print("Series + Series (with NaN):")
print(my_series + my_series2)

# Using .add() to handle the NaN in the first series
print("\nSeries + Series (using fill_value=0):")
print(my_series.add(my_series2, fill_value=0))
```

### String Arithmetic Example
A common use case for operators is converting numeric data into formatted strings (like currency). This involves casting to float, then to string, then adding the symbol.

```python
sales_series = pd.Series([0, 5, 155, 0, 518], name="Sales")

# Complex Operation: Convert to Currency String
currency_series = "$" + sales_series.astype("float").astype("string")

print("Currency Conversion:")
print(currency_series)
```

### Other Operations: Division and Multiplication
Note that dividing an integer series by a number will automatically cast the result to a **float64**.

```python
# Division
print("Division (Casts to float):")
print(my_series2 / 2)

# Floor Division (Returns nearest whole number)
print("\nFloor Division:")
print(my_series2 // 2)

# Order of Operations
print("\nOrder of Operations ((S1 + S2) / 2):")
print((my_series2 + my_series2) / 2)
```

### Practical Example: Data Cleaning
A common task is cleaning "price" strings (e.g., "$4.99") so they can be converted to numeric floats for analysis. If you try to convert "$4.99" directly, you will get an error.

```python
# Creating a series with string prices
prices = pd.Series(["$3.50", "$1.30", "$10.00", "$3.00"], name="Price")

# 1. Test if price contains '3'
print("Contains '3':")
print(prices.str.contains("3"))

# 2. Strip '$' and convert to float
clean_prices = prices.str.strip("$").astype("float")
print("\nCleaned Numeric Prices:")
print(clean_prices)
```

### Advanced String Slicing and Splitting

## 10. Missing Data Representation in pandas

Missing data is a common reality when reading files or databases. In pandas, this is primarily represented in two ways: **NumPy's `NaN`** and **pandas' `NA`**.

### 1. NumPy's `NaN` (Not a Number)
* **Placeholder:** Indicates no data was present in a row or column.
* **Efficiency:** More efficient than Python's `None` type.
* **Float Coercion:** `NaN` is treated as a float. This means if you have a column of integers and add a single `NaN`, the entire column will be converted (coerced) to floats.
* **Arithmetic Behavior:** Operations on `NaN` return `NaN` by default (e.g., `2 + NaN = NaN`). To bypass this, use pandas arithmetic methods with a `fill_value`.

### 2. pandas' `NA` (Nullable Integers)
* **Native Support:** Introduced in late 2020 to allow missing values in integer-based data without forcing a conversion to floats.
* **Sophisticated Types:** Uses pandas-native types (like `Int16` or `Int64` with a capital 'I') to support nullable values.
* **Current State:** While powerful for power users, it is still newer and can occasionally revert to `NaN` behavior during certain operations.

### Demo: `NaN` vs `NA` Behavior
Observe how the different missing value types affect the resulting Series data type.

```python
# Example 1: NumPy NaN (Coerces integers to floats)
sales_nan = pd.Series([10, 20, np.nan, 40, 50], name="Sales_NaN")
print("Series with np.nan:")
print(sales_nan)
print(f"Dtype: {sales_nan.dtype}")

# Example 2: pandas NA (Keeps integers as integers)
# Note the use of the capital 'I' in the dtype to allow nulls
sales_na = pd.Series([10, 20, pd.NA, 40, 50], name="Sales_NA", dtype="Int64")
print("\nSeries with pd.NA:")
print(sales_na)
print(f"Dtype: {sales_na.dtype}")
```

### Arithmetic with Missing Data
As discussed, standard operators will propagate the missing value. We use `.add()` to impute a value during the calculation.

```python
# Standard addition returns NaN
print("Standard Addition (2 + NaN):")
print(sales_nan + 2)

# Arithmetic method with fill_value allows calculation
print("\nAddition with fill_value=0:")
print(sales_nan.add(2, fill_value=0))
```

### Summary
For most analytical use cases, the distinction between `NaN` and `NA` isn't critical, as operations on numeric series work similarly regardless of whether the type is float or integer. However, identifying and summarizing these values is the next essential step in your data cleaning workflow.

## 11. Identifying Missing Data

Once we understand how missing data is represented (as `np.nan` or `pd.NA`), we need to be able to identify where those values are and summarize them. This is a critical step in the data cleaning process.

### The `.isna()` Method
The primary tool for detection is the `.isna()` method. It performs a boolean test across the entire Series:
* Returns **True** if a value is missing (NaN or NA).
* Returns **False** if the value is present.

Because this method returns a Boolean series, we can use it as a **mask** to filter our data or use it for quick mathematical summaries.

```python
# Creating a 'checklist' series with missing data
checklist = pd.Series(["Complete", np.nan, np.nan, np.nan, "Complete"], name="Checklist")

# 1. Identifying missing values with isna()
print("Boolean mask for missing data:")
print(checklist.isna())

# 2. Summarizing the count of missing values
# True = 1, False = 0; so .sum() gives the total count of missing rows
print(f"\nTotal missing values: {checklist.isna().sum()}")

# 3. Summarizing the percentage of missing values
print(f"Percentage of data missing: {checklist.isna().mean() * 100}%")
```

### Identifying Missing Data with `value_counts()`
By default, the `.value_counts()` method suppresses NA values—it won't show them in the results. To include them in your frequency summary, you must set the `dropna` parameter to `False`.

This is useful for measuring the frequency of missing data against existing categories in a single view.

```python
# Default value_counts() (Excludes NaNs)
print("Default value_counts:")
print(checklist.value_counts())

# Including missing values in the summary
print("\nValue counts including NaNs:")
print(checklist.value_counts(dropna=False))
```

### Summary
To summarize a column's missingness:
* Use `.isna().sum()` for a **raw count**.
* Use `.isna().mean()` for the **percentage**.
* Use `.value_counts(dropna=False)` to see the **distribution** relative to other categories.

## 12. Handling Missing Data

Handling missing data is a challenging problem. The correct approach—dropping rows, filling them with zeros, or using the mean—depends entirely on the specific data you are working with. 

### Main Methods for Handling Missing Data
* **`.dropna()`**: Removes any rows containing missing values from your Series.
* **`.fillna()`**: Replaces missing values with a specified value (e.g., 0, "Incomplete", or the mean).

### 1. Dropping Missing Data
When you use `.dropna()`, be aware that it creates gaps in your index. If you have a continuous index from 0 to 4 and drop the middle rows, your index will remain `0` and `4`. 

**Pro-Tip:** After dropping rows, you may want to use `.reset_index(drop=True)` to return to a continuous integer index.

```python
# Example: Dropping NaN values
checklist = pd.Series(["Complete", np.nan, np.nan, np.nan, "Complete"], name="Checklist")
dropped_series = checklist.dropna()

print("Series after dropna() (Note the index gaps):")
print(dropped_series)

# Resetting index to clean up gaps
print("\nSeries after reset_index(drop=True):")
print(dropped_series.reset_index(drop=True))
```

### 2. Filling Missing Data
Filling values is often safer than dropping them, especially for categorical data where you can simply create an "Incomplete" or "Missing" category. 

For numeric data, common strategies include:
* **Filling with 0**: Appropriate if the missing value represents an absence of activity (like zero sales).
* **Filling with the Mean**: Keeps summary statistics the same; very common in machine learning applications.



```python
# Numeric Series example
sales = pd.Series([10, pd.NA, pd.NA, 30, 40], name="Sales", dtype="Int64")

# 1. Filling with zero
print("Filled with 0:")
print(sales.fillna(0))

# 2. Filling with the mean
# FIX: nullable Int64 cannot store a fractional mean directly, so cast first.
print("\nFilled with the Mean:")
sales_float = sales.astype("Float64")
print(sales_float.fillna(sales_float.mean()))
```

### Summary
Always consult a data subject matter expert to understand *why* data is missing before fixing it. You might find that the missing data is due to a pipeline error that should be fixed at the source, rather than patched in pandas.

## 13. Applying Custom Functions to Series

The `.apply()` method allows you to run custom Python functions on every element of a pandas Series. While highly flexible, it is important to remember that these functions are **not vectorized**. 

### Efficiency Warning
* **Performance:** `.apply()` is significantly less efficient than native pandas or NumPy functions. 
* **Best Practice:** Always look for a built-in pandas method (like `.str` or `.where()`) before resorting to `.apply()`, especially for large datasets.
* **Use Case:** Use `.apply()` only for one-off tasks on small datasets or when the logic is so complex that recoding it in native pandas isn't worth the time.

### 1. Using Formal Functions
You can define a standard Python function and pass it to `.apply()`. The first argument of the function will automatically be the individual element from the Series.

```python
# Define a custom discount function
def apply_discount(price):
    if price > 20:
        return round(price * 0.9, 2)
    return price

wholesale_prices = pd.Series([15, 25, 10, 50, 18], name="Prices")

# Apply the custom function
discounted = wholesale_prices.apply(apply_discount)
print("Prices after custom discount function:")
print(discounted)
```

### 2. Using Lambda Functions
For quick, one-line manipulations, Lambda functions are more common than formal function definitions.

```python
# Extracting the last character of a string using a Lambda
string_series = pd.Series(["day 0", "day 2", "day 4"], name="Days")

extracted_ints = string_series.apply(lambda x: x[-1])
print("Extracted values via Lambda (Alternative to .str[-1]):")
print(extracted_ints)
```

### 3. Passing Additional Arguments
If your custom function requires more than one argument, you can pass them as a list or tuple to the `args` parameter in `.apply()`.

```python
# Function with multiple arguments
def find_char(string, char):
    if char in string:
        return "found it"
    return "nope"

# Applying with additional arguments
search_results = string_series.apply(find_char, args=("2",))
print("Search results using multiple arguments:")
print(search_results)
```

### Summary
`.apply()` is your "Swiss Army Knife"—it can do anything, but it's slower than specialized tools. Use it as a last resort when native vectorization isn't an option.

## 15. Dropping Columns and Rows

The `.drop()` method is essential for removing irrelevant data. Removing unnecessary columns reduces memory consumption and makes the remaining data much easier to manage.

### The Axis Parameter
To use `.drop()`, you must specify the **axis** to tell pandas whether you are targeting rows or columns:
* **`axis=0`**: Targets **Rows** (the default behavior).
* **`axis=1`**: Targets **Columns**.

### Inplace vs. Assignment
By default, `.drop()` **does not** change the original DataFrame; it returns a new, reduced copy. To modify the original data permanently, you have two options:
1. Use `inplace=True` (Note: This returns `None`, so you cannot chain it further).
2. Assign the result back to a variable (e.g., `df = df.drop(...)`).

### 1. Dropping Columns
Columns are dropped most often when they are redundant (like an ID column that mimics the index) or irrelevant to the specific analysis.

```python
oil_df = pd.read_csv("data/retail/oil.csv")

oil_df['euro_price'] = oil_df['price'] * 1.1
oil_df.head()
```

```python
# Now that we have the euro_price column, we can drop the original price column
oil_euro = oil_df.drop('price', axis=1)
oil_euro
```

### 2. Dropping Rows
Rows are dropped using `axis=0`. While less common than filtering based on logic, dropping by index is useful for removing specific outliers or ranges. 

**Warning:** Dropping rows results in a **non-consecutive index**. You may need to use `.reset_index(drop=True)` to fix this.

```python
oil_euro.drop(1, axis=0)
# Notice what happened to the index after dropping row 1
```

```python
oil_euro.drop(1, axis=0).reset_index()
```

### Summary
* **Memory Management:** Drop unnecessary columns as soon as you identify them to save resources.
* **Best Practice:** During development, assign the result to a new variable (e.g., `oil_euro`) rather than using `inplace=True`. This keeps your original data safe in case you need to reference a dropped column later.

## 16. Identifying and Dropping Duplicates

Occasionally when we read in data, we might find that there are a lot of duplicates in our data. Other times we might accidentally create duplicate rows by dropping a key ID column or by performing a bad join. Either way, sometimes duplicates can negatively impact the quality of an analysis, so being able to identify them is going to be helpful.

### Identifying Duplicates

* **The `.duplicated()` method**: This method will identify duplicate rows of data. 
* **How it works**: It doesn't consider the first time it sees a row to be a duplicate, but it will return `True` if any following rows are duplicates of a preceding row.
* **Column Specificity**: By specifying the `subset` argument, we can look for duplicates within a specific column of data as opposed to duplicates across all columns.

### Dropping Duplicates

The `.drop_duplicates()` method allows us to remove these rows. We can refine this using the following arguments:

| Argument | Description |
| :--- | :--- |
| **`subset`** | Judges a row to be a duplicate if a specific column (or list of columns) is repeated. |
| **`keep`** | Determines which instance to keep: `'first'` (default) or `'last'`. |
| **`ignore_index`** | Automatically resets the index to ensure there are no gaps (non-consecutive values) while dropping rows. |

### Demo: Identifying and Removing Duplicates

In this demonstration, we'll see how to spot duplicates by comparing the number of rows to the number of unique values, and then use the specialized methods to clean the data.

```python
# Creating a duplicate row at the end of the DataFrame
# Grabbing the last row and appending it to the end
retail_df = pd.concat([retail_df, retail_df.iloc[[-1]]], ignore_index=True)

print(f"Total Rows: {len(retail_df)}")
print(f"Unique Dates: {retail_df['date'].nunique()}")

# Identifying the duplicate row
print("\nSum of duplicated rows:")
print(retail_df.duplicated().sum())

# Identifying duplicates in a specific column (subset)
print("\nSum of duplicated prices (Subset='price'):")
print(retail_df.duplicated(subset='sales').sum())
```

### Refining the Drop Operation

Using `keep='last'` allows us to keep the final instance of a duplicate row, while `ignore_index=True` handles the index cleanup automatically.

```python
# Dropping duplicates while keeping the last instance and resetting the index
retail_cleaned = retail_df.drop_duplicates(
    subset='date', 
    keep='last', 
    ignore_index=True
)

print("Shape of cleaned data:")
print(retail_cleaned.shape)
retail_cleaned.tail()
```

### Summary
* **Spot Checking**: Compare row counts to `.nunique()` counts to see if values that should be unique (like IDs or Dates) are repeated.
* **Default Behavior**: `drop_duplicates` without arguments will remove exact row copies, leaving the original index gaps unless specified otherwise.
* **Index Management**: Use `ignore_index=True` to maintain a continuous set of integers after dropping.

## 17. Handling Missing Data in DataFrames

Working with missing data in DataFrames is very similar to Series, but we have additional options when dealing with multiple columns. Missing data in pandas is typically represented by the `pd.NA` or `np.NaN` types.

### Identifying Missing Data

* **The `.isna().sum()` method**: Calling this on an entire DataFrame returns a **Series** containing the count of missing values for each column.
* **The `.info()` method**: This method provides a summary of the DataFrame, including the count of non-null (non-missing) values. You can calculate the number of missing values by subtracting the non-null count from the total number of entries.

### Fixing Missing Data with `fillna()`

By default, `.fillna()` will replace all missing values in the entire DataFrame with a single specified value. However, it is often more useful to fill different columns with different values (e.g., filling a price with `0` but a product name with `"Unknown"`).

* **Dictionary Mapping**: You can pass a dictionary to `.fillna()` where the **keys** are column names and the **values** are the replacements for those specific columns.

### Dropping Missing Data with `dropna()`

* **Default Behavior**: By default, `.dropna()` will remove any row that has **at least one** missing value in any column. This can sometimes result in losing a significant amount of data.
* **The `subset` Argument**: This allows you to specify a single column (or list of columns) to check for missing values. Rows will only be dropped if they have missing data in the specified subset, leaving other rows intact even if they have missing values in other columns.

### Demo: Identifying and Filling Missing Data

In this demonstration, we'll use `isna().sum()` to identify problem areas and then use a dictionary with `fillna()` to apply column-specific fixes.

```python
# 1. Identifying missing values across the whole DataFrame
print("Missing values per column:")
print(oil_df.isna().sum())

# 2. Filling specific columns using a dictionary
# We want to fill price with 0 but product with 'missing'
fill_values = {'price': oil_df['price'].mean()}
oil_filled = oil_df.fillna(value=fill_values)

print("\nDataFrame after specific fillna:")
print(oil_filled.head())
```

### Refining the Drop Operation

Sometimes dropping all rows with any missing values is "overkill." We can use `subset` to target specific columns, such as unique IDs required for analysis.

```python
# Dropping rows only if 'price' is missing
oil_subset_drop = oil_df.dropna(subset=['price'])

print(f"Original row count: {len(oil_df)}")
print(f"Row count after targeted drop: {len(oil_subset_drop)}")
```

### Summary
* **`isna().sum()`**: The fastest way to get a column-wise count of missing data.
* **Targeted Fill**: Use dictionaries with `fillna()` to avoid applying nonsensical values (like numbers in text columns).
* **Targeted Drop**: Use the `subset` argument in `dropna()` to avoid losing too much data.

## 18. Renaming and Reordering Columns

Modifying the column axis is a common task when cleaning external data (like CSVs) to make column names more programmatic or intuitive for analysis.

### Renaming Columns
There are two primary ways to rename your columns:

1.  **Direct Assignment via `.columns`**: You can assign a list of new names directly to the columns property. The list must be in the exact same order as the existing columns. This is also useful for programmatic changes, like using a list comprehension to make all names uppercase.
2.  **The `.rename()` Method**: This method uses a dictionary where keys are existing names and values are the new names. 
    * **Not In Place**: By default, this returns a new DataFrame rather than modifying the original.
    * **Chaining**: Because it returns a DataFrame, it is convenient for chaining other methods like `.query()`.
    * **Lambda Support**: You can pass a Lambda function to the `columns` argument (e.g., `columns=str.upper`) to format all names at once.

### Reordering Columns
While you can sort columns alphabetically, you often need a specific custom order.

* **The `.reindex()` Method**: Use the `labels` argument to pass a list of column names in your desired order.
* **Axis Requirement**: You must specify `axis=1` (or `axis='columns'`) to ensure you are reordering the column axis rather than the row index.

### Demo: Renaming and Reordering

In this demonstration, we will practice both programmatic renaming and specific reordering of a product DataFrame.

```python
# Sample DataFrame
product_df = pd.DataFrame({
    "product": ["Coffee", "Tea"],
    "price": [3.50, 2.50],
    "product_id": [101, 102]
})

# 1. Renaming via assignment (Programmatic uppercase)
product_df.columns = [col.upper() for col in product_df.columns]
print("Direct Assignment (Upper):")
print(product_df.columns)

# 2. Renaming via .rename() method (Dictionary mapping)
# Note: Mapping back to specific names
product_df = product_df.rename(columns={"PRODUCT": "product_name", "PRICE": "cost"})
print("\nRenamed with Dictionary:")
print(product_df.head())

# 3. Reordering columns via .reindex()
new_order = ["product_id", "product_name", "cost"]
product_df = product_df.reindex(labels=new_order, axis=1)
print("\nReordered Columns:")
print(product_df.head())
```

### Summary
* **Programmatic Flexibility**: Use list comprehensions or Lambda functions within `.rename()` for bulk formatting.
* **Method Chaining**: Use `.rename()` when you want to keep the original DataFrame intact or chain operations together.
* **Correct Axis**: Always remember `axis=1` when reindexing columns.

## 22. The Categorical Data Type

The final portion of our DataFrame section focuses on **memory optimization**. Understanding how pandas handles data types can significantly reduce the amount of RAM your analysis requires.

### Memory Optimization Basics
* **Numeric Types**: Standard NumPy integers and floats default to 64-bit values. Pandas-native versions (capitalized, like `Int64`) allow for nullable values (`pd.NA`) without forcing a conversion to floats, which are more memory-intensive than integers.
* **Object Type**: This is a broad, flexible data type capable of storing any Python object (strings, lists, images), but it is often inefficient for text.
* **Categorical Type**: This maps text data with repeated values to a numeric array for high efficiency.

### How Categorical Data Works
If a dataset has 10,000 rows but only 5 unique text categories, the `object` type stores the full text for every row. The `categorical` type instead creates a dictionary to map those categories to integers (e.g., `Beverages: 0`, `Dairy: 1`) and stores those small integers in the DataFrame.

**Rule of Thumb**: Only use the categorical data type when the number of unique categories is less than **half the number of total rows** (Unique Categories < Rows / 2).

### Benefits
* **Huge Memory Savings**: Can reduce DataFrame size by 50% or more.
* **Seamless Display**: Even though data is stored as integers on the backend, pandas still displays the original text values for convenience.

### Demo: Optimizing Memory with Categories

In this demonstration, we use a retail dataset with ~3 million rows to show how converting a single column can drastically lower RAM usage.

```python
# 1. Load a large dataset and check memory usage
# 'deep' provides a realistic read of RAM usage for object types
retail = pd.read_csv('data/retail/retail_2016_2017.csv')
print("Original Memory Usage:")
print(retail.info(memory_usage='deep'))

# 2. Convert 'family' from object to category
retail['family'] = retail['family'].astype('category')

# 3. Verify the conversion and check new memory usage
print("\nNew Memory Usage after Categorical Conversion:")
print(retail.info(memory_usage='deep'))

# 4. Confirm data still displays normally
print("\nValue Counts (displayed as text):")
print(retail['family'].value_counts().head())
```

### Summary
* **Check your Info**: Always use `info(memory_usage='deep')` to see the true cost of your data.
* **Vectorization**: Categorical data types are highly efficient for data processing.
* **Quirks**: Be aware that some complex data manipulations may occasionally force categorical data back into the `object` type.

* **Values**: The target data types for those specific columns.

### 3. Compatibility and Cleaning
Just like with Series, the target data type must be compatible with the current data. For example, a string column containing a dollar sign (e.g., "$10.00") cannot be converted directly to a float; it will raise an error. 

**The `assign()` Method for Cleaning:**
The `.assign()` method is an excellent tool for type conversion when multiple cleaning steps are required. You can clean the data (like removing special characters) and perform the conversion in a single, readable chain.

### Demo: Multi-Column and Cleaned Conversions

In this demonstration, we'll convert multiple columns using a dictionary and show how to use `assign()` to fix an incompatible numeric column.

```python
import pandas as pd

# 1. Multi-Column Conversion using a Dictionary
# FIX: the original notebook referenced an undefined df and columns that do not exist in the uploaded files.
# Build a tiny self-contained demo dataframe for dtype conversion.
conversion_dict = {
    'date': 'datetime64[ns]',
    'onpromotion': 'float64'
}

df = pd.DataFrame({
    'date': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'onpromotion': [1, 0, 3],
    'price': ['$10.50', '$7.25', '$12.00']
})

df = df.astype(conversion_dict)
print("Dtypes after multi-column conversion:")
print(df.dtypes)

# 2. Using .assign() for Cleaning + Conversion
df = df.assign(
    price=df['price'].str.replace('$', '', regex=False).astype('float')
)

print("\nCleaned Price Column:")
print(df['price'].head())
```

## 24. Pro Tip: Memory Usage and Optimization

Memory optimization is vital when working with large datasets because pandas DataFrames are stored entirely in RAM. Reducing memory consumption creates a more efficient workflow and prevents system crashes when datasets approach the limits of your hardware.

### Core Optimization Strategies
* **Drop Unnecessary Columns**: Remove columns that add no value or are redundant (e.g., an ID column identical to the index).
* **Convert Object Types**: Cast object columns (strings) to numeric or datetime types whenever possible. 
* **Downcasting**: Reduce numeric data to the smallest appropriate bit size.
* **Categorical Types**: Use the `category` type when unique values are less than half the total number of rows.

### Monitoring Memory
To get an accurate reading of memory consumption, use the `.memory_usage()` method or `.info()` with the `deep=True` parameter. This ensures that the memory used by the actual data within object columns is fully accounted for.

### Demo: Reducing Memory Footprint

In this demonstration, we'll see how dropping a redundant ID column and converting date strings to proper `datetime64` objects can significantly reduce the size of a DataFrame.

```python
import pandas as pd
import numpy as np

# FIX: continue from a self-contained classroom dataframe instead of an undefined df.
df = pd.DataFrame({
    'id': [1, 2, 3],
    'start_date': ['2024-01-01', '2024-01-08', '2024-01-15'],
    'students_enrolled': ['25', '-', '40'],
    'class_level': [8, 10, 12],
    'price': [199, 299, 399]
})

# 1. Initial Memory Check
print("Initial Memory Usage by Column:")
print(df.memory_usage(deep=True))
print(f"\nTotal Memory: {df.memory_usage(deep=True).sum()} bytes")

# 2. Drop Redundant Columns
df = df.drop(columns=['id'])

# 3. Convert Object to Datetime
df['start_date'] = pd.to_datetime(df['start_date'])

# 4. Cleaning and Converting Numeric Objects
df['students_enrolled'] = df['students_enrolled'].replace('-', np.nan).astype('Float64')

print("\nDtypes after cleaning:")
print(df.dtypes)
print(f"\nOptimized Memory So Far: {df.memory_usage(deep=True).sum()} bytes")
```

### Summary
* **Scale Matters**: While these steps save bytes on small DataFrames, the impact scales massively as datasets grow.
* **Pre-cleaning**: Always check for non-standard missing value representations (like hyphens) that might force numeric data into object types.
* **RAM Requirements**: Aim to have 5–10x more RAM than your DataFrame size for smooth processing.

## 25. Pro Tip: Downcasting Numeric Data Types

By default, pandas assigns 64-bit data types to integers and floats because it does not know the maximum potential value of a column. However, if you know the range of your data, you can "downcast" these to smaller bit sizes to save significant memory.

### Numeric Bit Ranges
Understanding the limits of each bit size is key to safe downcasting:
* **8-bit (int8)**: Ranges from -128 to 127.
* **16-bit (int16)**: Ranges from -32,768 to 32,767.
* **32-bit (int32)**: Ranges from roughly -2 billion to 2 billion.
* **64-bit (int64)**: Capable of handling almost any number imaginable.

### Optimization Impact
In our demonstration, combining column dropping, type conversion, and numeric downcasting reduced memory consumption by approximately **63%**. While individual savings may seem small (in bytes), these efficiencies scale massively when applied to millions of rows.

**Note on Categorical Overhead**: The categorical data type only adds value when there are many repeated values. If the number of unique categories is nearly equal to the number of rows, converting to `category` can actually *increase* memory usage due to the additional overhead.

### Demo: Downcasting in Action

We will now cast our numeric columns to the smallest safe bit sizes based on our knowledge of the data ranges (e.g., class levels never exceeding 12).

```python
# 1. Downcasting based on known ranges
# Int8 is safe for 'class_level' (max 12)
# Int16 is safe for 'price' (max 599) and 'students_enrolled'
df = df.astype({
    'class_level': 'Int8',
    'price': 'Int16',
    'students_enrolled': 'Int16'
})

print("Dtypes after downcasting:")
print(df.dtypes)

print(f"\nFinal Optimized Memory: {df.memory_usage(deep=True).sum()} bytes")
```
